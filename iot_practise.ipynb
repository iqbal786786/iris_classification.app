```python

student = []
for i in range(int(input())):
    name = input()
    score = float(input())
    student.append([name,score])

student_sublist = sorted(student, key= lambda x: x[1])

second_max = student_sublist[-2][1]

sorted_answer = []
for i in range(len(student_sublist)):
    if second_max == student_sublist[i][1]:
        sorted_answer.append(student_sublist[i])

sorted_answer = sorted(sorted_answer, key= lambda x:x[0])
for i in range(len(sorted_answer)):
    print(sorted_answer[i][0])
```
